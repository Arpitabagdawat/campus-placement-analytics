**step-4**
**EDA — Placed vs Not Placed**

**Business Question 1**

```python
# Overall placement count
df['status'].value_counts()
```

```python
#Placement percentage:
placement_rate = df['status'].value_counts(normalize=True) * 100
placement_rate
```

**Business Question 2**

```python
#Gender-wise placement rate
pd.crosstab(df['gender'], df['status'], normalize='index') * 100
```

**Skill Gap – Academics vs Placement**

```python
df.groupby('status')[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']].mean()
```

```python
pd.crosstab(df['workex'], df['status'], normalize='index') * 100
```

**Salary Analysis (Placed students)**

```python
# Salary stats (only placed students)
df[df['status'] == 'Placed']['salary'].describe()
```

```python
#Salary by Specialisation:
df[df['status'] == 'Placed'].groupby('specialisation')['salary'].mean()
```

**Final Insights:**

**1. Overall placement rate is ~69%.**

**2. Male students have slightly higher placement rate than females.**
    
**3. Higher academic scores increase chances of placement.**
    
**4. Work experience greatly improves placement probability.**

**5. Average salary is around 2.8 LPA (max up to 9.4 LPA).**

**6. Mkt&Fin students earn more on average than Mkt&HR.**
